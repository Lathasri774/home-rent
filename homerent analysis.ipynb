```python
import matplotlib.pyplot as plt
values=[100,30,40,25,90,200]
lab=["home","power","water","vehicles","food","others"]
plt.pie(values,labels=lab)
plt.axis("equal")
plt.show()
plt.pie(values,labels=lab,shadow=True)
plt.axis("equal")
plt.show()
plt.pie(values,labels=lab,autopct='%1.1f%%',radius=2)

plt.show()
```
